# Problem Statement:
To predict the percentage of the students based on the hours of study using the Simple Regression Model.

```python
##Importing the necessary packages
import numpy as np
import pandas as pd
import plotly.express as px
```

```python
# importing the neccessary data
df = pd.read_csv('student_scores - student_scores.csv')
```

```python
df.head()
```

```python
def plot_scatter(t):
       fig=px.scatter(x = t['Hours'],y = t['Scores'],width = 500,height =400,labels={'x': 'Hours', 'y': 'Scores'})
       
       return fig.show()
```

```python
plot_scatter(df)
```

Linear relationship can be observed between the two variables.

Lets pepare the model

```python
X = df.iloc[:, :-1].values  
y = df.iloc[:, 1].values  
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test, y_train, y_test = train_test_split(X,y,random_state=0)
```

```python
y
```

Now fit the prepared model

```python
from sklearn.linear_model import LinearRegression
Lr = LinearRegression()
model=Lr.fit(X_train, y_train)
```

Making Predictions

```python
y_pred = Lr.predict(X_test)
```

```python
print(X_test),print(y_pred)
```

```python
def score_predict(hours):
    hours  = np.array(hours)
    own_pred = Lr.predict(hours.reshape(1,-1))
    if own_pred > 100:
        print("Error")
    else:
        return   print("No. of hours studied :",hours) , print("Percentage expected:", own_pred[0])
```
